```python
import anndata
```

```python
adata = anndata.read_h5ad("../data/processed/anndata.h5ad")
```

```python
# adata.X # the genes
```

```python
# adata.obs # shows metadata
```

```python
df_X_with_obs = adata.to_df().join(adata.obs)
```

```python
df_X_with_obs.sample(5)
```

```python
df_X_with_obs['cell_label'] = df_X_with_obs.index
```

```python
df_X_with_obs.reset_index()
```

```python
df_X_with_obs["target"] = (
    df_X_with_obs["sample"]
    .apply(lambda x: (x.split("_")[-1])[0]) # split "sample" column and get the last part containing Cancer1, EMS, N-5 Norm2, etc.... [0] gets the first letter C, E, N
    .map({"C": "Cancer", "E": "EMS", "N": "Normal"})
)

set(df_X_with_obs.target)
```

```python
df_X_with_obs.to_parquet("../data/processed/df_X_with_obs.parquet")
```

```python
columns_to_drop = [
    "cell_label",
    "sample",
    "batch",
    "n_genes",
    "n_genes_by_counts",
    "total_counts",
    "total_counts_mt",
    "pct_counts_mt",
]

df_X_with_target = df_X_with_obs.loc[
        :, ~df_X_with_obs.columns.isin(columns_to_drop)
        ].copy()
```

```python
df_X_with_target.sample(4)
```

```python
df_X_with_target.to_parquet("../data/processed/df_X_with_target.parquet")
```
